### Visualizing

```python
# It counts the occurence of all individuals years in the date column
df['date_n'].dt.year.value_counts()
```

```python
# Distribution of data across different years(which years get more released movies and tv shows)


# Count occurrences of each year and convert to DataFrame
year_counts = df['date_n'].dt.year.value_counts().sort_index().reset_index()
year_counts.columns = ['Year', 'Count']

# Create the bar plot using Seaborn
plt.figure(figsize=(12, 6))  # Adjust figure size
sns.barplot(x='Year', y='Count', data=year_counts, palette='deep')

# Labels and title
plt.xlabel('Release Year')
plt.ylabel('Number of Movies & TV Shows')
plt.title('Distribution of Releases Across Different Years')

# Rotate x-axis labels if there are many years
plt.xticks(rotation=45)
plt.show()
```
